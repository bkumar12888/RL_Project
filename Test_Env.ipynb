```python
# Importing libraries
import numpy as np
import random
import math
from collections import deque
import collections
import pickle
import operator
```

```python
# Import the environment
from Env import CabDriver
env = CabDriver()
```

```python
# Loading the time matrix provided
Time_matrix = np.load("TM.npy")
```

```python
action_space, state_space, state = env.reset()
print(action_space)
print(state_space)
print(state)
```

```python
requests = env.requests(state)
print("REQUESTS: {}".format(requests))
```

```python
rewards = []
for req in requests[1]:
    r =  env.reward_func(state, req, Time_matrix)
    rewards.append(r)

print(rewards)
```

```python
new_states = []
for req in requests[1]:
    s = env.next_state_func(state, req, Time_matrix)
    new_states.append(s)

print(new_states)
```

```python
n = 100 #number of rides taken by cab driver

for i in range(1, n):
    print("\n.................RIDE# {} ...............\n".format(i))
    
    requests = env.requests(state)
    print("REQUESTS: {}".format(requests))

    rewards = []
    for req in requests[1]:
        r =  env.reward_func(state, req, Time_matrix)
        rewards.append(r)

    print("REWARDS: {}".format(rewards))

    new_states = []
    for req in requests[1]:
        s = env.next_state_func(state, req, Time_matrix)
        new_states.append(s)
    print("NEW POSSIBLE STATES: {}".format(new_states))
    
    index, max_reward = max(enumerate(rewards), key=operator.itemgetter(1))
    state = new_states[index]
    print("MAXIMUM REWARD: {}".format(max_reward))
    print("NEW STATE: {}".format(state))
```
